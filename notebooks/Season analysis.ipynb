```python
import pandas as pd
import numpy as np
from bson import ObjectId

from betlib.models.game import find_game, update_game
from betlib.mongo import get_db

client = get_db("prono", mongolab=True)
```

```python
# Load parameters

seasons_path = "/Users/valentin/Projets/prono_foot/data/seasons"

csv_name = "55908dfe9734042e6bbd4360_2015_2016.csv"
```

```python
name_params = csv_name.replace(".csv", "").split("_")

division_id = name_params[0]
years = [name_params[1], name_params[2]]
```

```python
season = pd.read_csv(seasons_path + '/' + csv_name, sep=";")
season[season.index > 50]
```

```python
from itertools import groupby as g
def most_common(L):
  return max(g(sorted(L)), key=lambda(x, v):(len(list(v)),-L.index(x)))[0]
```

```python
# Nombre d'équipes total
teams = season["team_H"].unique()
teams_total = len(teams)
assert teams_total == len(season["team_A"].unique())

# Nombre de match par équipe
assert len(set([len(season[season["team_H"] == team]) for team in teams])) == 1
assert len(set([len(season[season["team_A"] == team]) for team in teams])) == 1
assert most_common([len(season[season["team_A"] == team]) for team in teams]) == most_common([len(season[season["team_H"] == team]) for team in teams])
```
