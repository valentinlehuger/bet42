### Analyse crawls

```python
from betlib.models.team import add_team
from betlib.utils import normalize
```

```python
country = "france"
championship = "ligue_1"
years = [2014, 2015]
season_params = "%s_%s_%s" % (championship, years[0], years[1])
```

```python
with open("/Users/valentin/Projets/prono_foot/resources/%s.json" % season_params) as season_file:
    season = json.load(season_file)
```

```python
teams = set()

for day in season:
    for game in season[day]:
        game["team_H"] = normalize(game["team_H"])
        game["team_A"] = normalize(game["team_A"])
#        teams.add(game["team_H"])
#        teams.add(game["team_A"])
```

```python
for team in list(teams):
    db_team = {
        "country": country,
        "aliases": [],
        "name": team
    }

    add_team(db_team)
```
